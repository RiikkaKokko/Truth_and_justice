```python
from pathlib import Path
from estnltk import Text
import os
```

# Lemmatization

```python
#Go through all books
books=[]
for filename in os.listdir('Data/Books/'):
    if filename.startswith("tõde_ja_õigus_wo"):
        books.append(filename)
print(books)
```

```python
#ei õnnestu, tekib error, salvestab ainult esimese raamatu
count=0
uus_list_of_lemmas=[]
for book in books:
    count+=1
    text=Text((Path('Data/Books')/book).read_text().lower())
    text.analyse('morphology')
    lemmas=text.lemma
    list_of_lemmas=list(text.lemma)
    uus_list_of_lemmas=[elem[0] for elem in list_of_lemmas]
    print(uus_list_of_lemmas[:2000])
    filename='tõde_ja_õigus_lemmas' + str(count) +'.txt'
    save_book=open(Path('Data/Books')/filename, "w")
    save_book.write(str(uus_list_of_lemmas))
    save_book.close()  
```

```python
def save_lemmas_file(path, filename):
    text=Text(Path(path).read_text().lower())
    text.analyse('morphology')
    lemmas=text.lemma
    list_of_lemmas=list(text.lemma)
    uus_list_of_lemmas=[elem[0] for elem in list_of_lemmas]
    filename=filename
    save_book=open(Path('Data/Books')/filename, "w")
    save_book.write(str(uus_list_of_lemmas))
    save_book.close()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator
stopwords=Path('Data/Estonian_Stopwords.txt').read_text().split('\n')
def create_word_cloud(string):
    maskArray = np.array(Image.open("Data/Wordclouds/estonian_flag.jpg"))
    mask_colors= ImageColorGenerator(maskArray)
    cloud = WordCloud(background_color = "white", max_words = 2000, max_font_size=256, 
                      random_state=15, mask = maskArray, stopwords = stopwords, normalize_plurals=False, color_func=mask_colors)
    cloud.generate(string)
    plt.show()
    cloud.to_file("wordCloud_lemma1.png")
create_word_cloud(tekst)
```
